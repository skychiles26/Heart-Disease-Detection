```python
print(df["EKG results"].value_counts(),
df["Slope of ST"].value_counts(),
df["Number of vessels fluro"].value_counts())
df['Heart Disease'] = df['Heart Disease'].replace(['Presence', 'Absence'], [1, 0])
df
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6,10,11])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(x_train)
sc.fit_transform(x_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=25, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units=25, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units=25, activation = 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation = 'sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = 'accuracy', )
```

```python
ann.fit(x_train, y_train, batch_size = 25, epochs = 200)
```
